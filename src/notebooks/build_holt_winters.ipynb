```python
# Import necessary libraries
import pandas as pd
from src.holt_winters import create_holt_winters_model
```

```python
# Load data from csv file
df = pd.read_csv("datasets/nasa_requests_per_minute.csv")
df['time'] = pd.to_datetime(df['time'])
df = df.set_index('time')

print(f"Loaded {len(df)} rows")
print(f"Date range: {df.index.min()} - {df.index.max()}")
df.head()
```

```python
# Split data into train, validation and test sets 70/15/15
total_entries = len(df)
train_size = int(total_entries * 0.7)
validation_size = int(total_entries * 0.15)

train_dataset = df.iloc[:train_size]
validation_dataset = df.iloc[train_size:train_size + validation_size]
test_dataset = df.iloc[train_size + validation_size:]
```

```python
# Train the model
model = create_holt_winters_model(
        seasonal_periods=60,  # Adjust based on your EDA
    )
    
model.fit(train_dataset['number_of_requests'].tolist())
    
 # Predict on validation set
predictions = []
for i in range(len(validation_dataset)):
    pred = model.predict(steps=1)
    predictions.append(pred)
```
